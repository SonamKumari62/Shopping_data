```python
df['seller_name'] = df['seller_name'].fillna('Unknown Seller')
```

```python
df['discount'] = df['discount'].fillna(df['discount'].median())

df['what_customers_said'] = df['what_customers_said'].fillna('No Review')

df['seller_name'] = df['seller_name'].fillna('Unknown Seller')

df['videos'] = df['videos'].fillna('No Video')

df['seller_information'] = df['seller_information'].fillna('Information Not Available')

df['variations'] = df['variations'].fillna('No Variations')
```

```python
df.isnull().sum()
```

```python
df[['title', 'rating', 'final_price', 'category']].head()
```

```python
high_rated = df[df['rating'] > 4]

high_rated.head()
```

```python
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
```

```python
expensive_products = df[df['final_price'] > 1000]

expensive_products.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['initial_price'] = df['initial_price'].fillna(0)
df['final_price'] = df['final_price'].fillna(0)

df['discount_amount'] = df['initial_price'] - df['final_price']
```

```python
df['discount_amount'] = df['initial_price'] - df['final_price']

df[['initial_price', 'final_price', 'discount_amount']].head()
```

```python
df['discount_percentage'] = (
    (df['initial_price'] - df['final_price'])
    / df['initial_price']
) * 100

df[['initial_price','final_price','discount_percentage']].head()
```

```python
df.to_csv('cleaned_ecommerce_products.csv', index=False)

print("Dataset saved successfully!")
```
